```python
file_path = 'Final_Filtered_Merged_Temperature_and_GHG_Data.csv'
data = pd.read_csv(file_path)

# Filter the data for Japan
japan_data = data[data['Country'] == 'Singapore']

# Select relevant columns
japan_data = japan_data[['Year', 'Annual Mean', 'GHG']]

# Calculate the correlation coefficient
correlation_coefficient = japan_data['Annual Mean'].corr(japan_data['GHG'])

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=japan_data, x='Annual Mean', y='GHG')
sns.regplot(data=japan_data, x='Annual Mean', y='GHG', scatter=False, color='red')
plt.title(f'Correlation between Temperature and GHG Emissions in Niger relation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('Annual Mean Temperature (Â°C)')
plt.ylabel('GHG Emissions(MtCO2e)')
plt.grid(True)
plt.show()

# Print the correlation coefficient
print(f'Correlation Coefficient: {correlation_coefficient:.2f}')
```

```python
import pandas as pd

# Load the data from the CSV file
file_path = 'Final_Filtered_Merged_Temperature_and_GHG_Data.csv'
data = pd.read_csv(file_path)

# Group the data by country and calculate the correlation coefficient for each country
correlation_by_country = data.groupby('Country').apply(lambda x: x['Annual Mean'].corr(x['GHG']))

# Loop through all countries and print their correlation coefficients
for country, correlation in correlation_by_country.items():
    print(f"Country: {country}, Correlation Coefficient: {correlation}")
```
